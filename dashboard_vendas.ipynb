# Construindo um Dashboard de Vendas com Streamlit

Neste projeto, explorei como transformar análises de dados em um dashboard interativo que pode ser usado por outras pessoas da empresa.

## O que é o Streamlit?

Streamlit é uma biblioteca Python que permite criar aplicações web de forma rápida, sem precisar conhecer HTML, CSS ou JavaScript. É muito usada para criar dashboards e relatórios interativos.

## 1. Carregando os Dados

````python
import pandas as pd

df = pd.read_excel('dataset_completo_editado.xlsx')
df.head()
````

````python
# Período dos dados
print(f"De: {df['Data'].min().strftime('%d/%m/%Y')}")
print(f"Até: {df['Data'].max().strftime('%d/%m/%Y')}")
````

## 2. Calculando Indicadores (KPIs)

KPIs são números importantes que resumem o desempenho do negócio.

````python
# Faturamento total
faturamento = df['Valor_Total'].sum()
print(f"Faturamento Total: R$ {faturamento:,.2f}")
````

````python
# Quantidade de itens vendidos
itens = df['Quantidade'].sum()
print(f"Itens Vendidos: {itens:,}")
````

````python
# Margem média de lucro
margem = df['Margem_Lucro'].mean()
print(f"Margem Média: {margem:.1f}")
````

````python
# Número de transações
transacoes = df.shape[0]
print(f"Nº de Transações: {transacoes}")
````

## 3. Criando Filtros

Para o dashboard ser útil, é importante que o usuário possa filtrar os dados.

````python
# Categorias disponíveis
categorias = df['Categoria'].unique()
print(f"Categorias: {categorias}")
````

````python
# Cidades disponíveis
cidades = df['Cidade'].unique()
print(f"Cidades: {cidades}")
````

````python
# Exemplo de filtragem
df_filtrado = df[
    (df['Categoria'].isin(['Eletronicos', 'Moveis'])) &
    (df['Cidade'] == 'Sao Paulo')
]
print(f"Registros encontrados: {len(df_filtrado)}")
````

## 4. Preparando os Gráficos

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Faturamento por categoria
faturamento_categoria = df.groupby('Categoria')['Valor_Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(x=faturamento_categoria.index, y=faturamento_categoria.values, palette='viridis')
plt.title('Faturamento por Categoria')
plt.ylabel('Valor Total (R$)')
plt.xlabel('')
plt.tight_layout()
plt.show()
````

````python
# Distribuição da margem de lucro
plt.figure(figsize=(8, 4))
sns.histplot(df['Margem_Lucro'] * 100, bins=10, kde=True, color='green')
plt.title('Distribuição da Margem de Lucro')
plt.xlabel('Margem (%)')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
````

````python
# Top 5 produtos mais vendidos
top_produtos = df.groupby('Produto')['Valor_Total'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.barplot(x=top_produtos.values, y=top_produtos.index, palette='viridis')
plt.title('Top 5 Produtos Mais Vendidos')
plt.xlabel('Faturamento (R$)')
plt.tight_layout()
plt.show()
````

## 5. O Dashboard Final

Depois de entender como calcular os indicadores e criar os gráficos, montei o dashboard interativo usando Streamlit.

O arquivo `dashboard_vendas.py` contém o código completo que:

- Carrega os dados de forma otimizada
- Mostra KPIs principais no topo
- Permite filtrar por categoria e cidade
- Exibe gráficos interativos
- Possibilita exportar os dados filtrados em Excel

Para executar o dashboard:

```bash
streamlit run dashboard_vendas.py
```

## 6. O que Aprendi

**Sobre Streamlit:**
- `st.title()` e `st.markdown()` para textos
- `st.columns()` para organizar em colunas
- `st.metric()` para mostrar indicadores
- `st.pills()` para filtros rápidos
- `st.pyplot()` para exibir gráficos do matplotlib
- `st.download_button()` para permitir download de arquivos

**Sobre performance:**
- `@st.cache_data` evita recarregar os dados a cada interação

**Sobre organização:**
- Separar a lógica de dados da visualização
- Tratar casos onde o filtro resulta em dados vazios
